```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.geometry import Point
```

```python
df_train = pd.read_csv("../data/raw/train_staiton.csv")
df_postcode = pd.read_csv("../data/raw/property_postcode.csv")
sf_postcode = gpd.read_file("../data/raw/VMADMIN/POSTCODE_POLYGON.shp")
```

```python
df_train['postcode'] = 0
df_train
```

```python
type(df_train['geometry'])
```

```python
for i in range(0, len(df_train)):
    point = Point(df_train['LONGITUDE'][i], df_train['LATITUDE'][i])
    for j in range(0, len(sf_postcode)):
        polygon = sf_postcode['geometry'][j]
        if point.within(polygon):
            df_train['postcode'][i] = sf_postcode['POSTCODE'][j]
            break
```

```python
df_train.to_csv('../data/raw/train_with_postcode.csv', index=False)
```
