```python
import pandas as pd
```

```python
df_high_school_rank = pd.read_csv('../data/raw/school_rank_more_info.csv')
```

```python
df_high_school_rank
```

```python
dic_postcode = df_high_school_rank.groupby('Postal_Postcode').count().to_dict()['School_Name']
dic_postcode
```

```python
df_sorted_high_school_rank = df_high_school_rank.sort_values(by=['Postal_Postcode'], ascending=True, ignore_index = True)
df_sorted_high_school_rank
```

```python
lst_postcode = sorted(list(set(df_high_school_rank['Postal_Postcode'])))
len(lst_postcode)
```

```python
headers = ["POSTCODE", "1-10", "11-50", "51-100", "101-150", "150+"]
high_school = pd.DataFrame(columns = headers, index=range(len(lst_postcode)))
```

```python
index = 0
for j in range(0, len(dic_postcode.keys())):
    postcode = list(dic_postcode.keys())[j]
    high_school.loc[j]["POSTCODE"] = postcode
    high_school.loc[j]["1-10"] = 0
    high_school.loc[j]["11-50"] = 0 
    high_school.loc[j]["51-100"] = 0 
    high_school.loc[j]["101-150"] = 0
    high_school.loc[j]["150+"] = 0 
    count = dic_postcode[postcode]
    for i in range(index, index+count):
        #print(df_sorted_high_school_rank.loc[i])
        if df_sorted_high_school_rank.loc[i]['rank'] <= 10:
            high_school.loc[j]["1-10"] += 1
        if (df_sorted_high_school_rank.loc[i]['rank'] > 10) & (df_sorted_high_school_rank.loc[i]['rank'] <= 50):
            high_school.loc[j]["11-50"] += 1
        if (df_sorted_high_school_rank.loc[i]['rank'] > 50) & (df_sorted_high_school_rank.loc[i]['rank'] <= 100):
            high_school.loc[j]["51-100"] += 1  
        if (df_sorted_high_school_rank.loc[i]['rank'] > 100) & (df_sorted_high_school_rank.loc[i]['rank'] <= 150):
            high_school.loc[j]["101-150"] += 1
        if (df_sorted_high_school_rank.loc[i]['rank'] > 150):    
            high_school.loc[j]["150+"] += 1
    index = index+count
```

```python
high_school
```

```python
high_school.to_csv('../data/raw/postcode_high_school.csv', index=False)
```

```python
headers = ["POSTCODE", "average_rank"]
high_school_ave = pd.DataFrame(columns = headers, index=range(len(lst_postcode)))
```

```python
index = 0
for j in range(0, len(dic_postcode.keys())):
    postcode = list(dic_postcode.keys())[j]
    high_school_ave.loc[j]["POSTCODE"] = postcode
    high_school_ave.loc[j]["average_rank"] = 0
    count = dic_postcode[postcode]
    for i in range(index, index+count):
        high_school_ave.loc[j]["average_rank"] += df_sorted_high_school_rank.loc[i]['rank']
    high_school_ave.loc[j]["average_rank"] = high_school_ave.loc[j]["average_rank"] / count
    index = index+count
```

```python
high_school_ave 
```

```python
high_school_ave.to_csv('../data/raw/high_school_ave.csv', index=False)
```
