```python
import pandas as pd 
```

```python
# download the data
from urllib.request import urlretrieve
urlretrieve("https://www.education.vic.gov.au/Documents/about/research/datavic/dv331_schoollocations2022.csv", 
            "school locations.csv")
```

```python
# read the data
df_school = pd.read_csv("school locations.csv", encoding='cp1252')
```

```python
df_school 
```

```python
# see how many categories are in the Education_Sector columns
df_school['Education_Sector'].groupby(df_school['Education_Sector']).count()
```

```python
# get the number of scholls in each postcode area
df_school['num_of_school'] = df_school['Postal_Postcode'].groupby(df_school['Postal_Postcode']).transform('count')
df_school
```

```python
# rename the columns and remove duplicated rows
df_school_postcode = df_school[['Postal_Postcode', 'num_of_school']]
df_school_postcode.rename(columns={'Postal_Postcode': 'POSTCODE'}, inplace=True)
df_school_postcode = df_school_postcode.drop_duplicates()
df_school_postcode
```

```python
# save the data
df_school_postcode.to_csv("school_postcode.csv")
```
