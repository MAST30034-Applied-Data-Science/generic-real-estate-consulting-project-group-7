```python
import pandas as pd
```

```python
# read the csv file of all schools and the school rank
df_school = pd.read_csv("../data/raw/school_remove_selective.csv", encoding='cp1252')
df_school_rank = pd.read_csv('../data/raw/school_rank.csv')
```

```python
# rename the column in school rank so that they have a same column to join on
df_school_rank.rename(columns={'School': 'School_Name'}, inplace=True)
pd.merge(df_school_rank, df_school, on='School_Name', how='inner')
```

```python
# select the useful columns
school_rank_more_info = pd.merge(df_school_rank, df_school, on='School_Name', how='inner')[['rank', 'School_Name', 'Location', 'MedianVCE Score', 'VCE40+%', 'Education_Sector', 'School_Type', 'Postal_Postcode','X', 'Y']]
```

```python
school_rank_more_info['School_Type'].groupby(school_rank_more_info['School_Type']).count()
```

```python
# save to school_rank_more_info.csv
school_rank_more_info.to_csv('../data/raw/school_rank_more_info.csv', index=False)
```
